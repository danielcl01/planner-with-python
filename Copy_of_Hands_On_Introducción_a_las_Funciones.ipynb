## **FUNCIÓN INTEGRADA**

## **FUNCIÓN DEFINIDA POR EL USUARIO**

Una **función definida por el usuario** es creada de manera personalizada para realizar tareas específicas y reutilizables en todo el programa. Este tipo de función se crea utilizando la palabra clave ` def` seguida del nombre de la función y los parámetros, si los hay.

Una función definida por el usuario tiene la siguiente estructura:

```
def nombre_funcion(parametro1, parametro2, ...):
    # Código que realiza una tarea específica

    # Devolver un valor (opcional)
    return valor
```

### **EJEMPLO 1:**

Función que calcula el cuadrado de un número:

````python
# difinir función

num_cuad = int(input('Ingresa un número: '))

def cuadrado(numero):
    return numero ** 2

cuadrado(num_cuad)
````

````python
# difinir función

num_par = int(input('Ingresa un número: '))

def par(numero):
    if numero % 2 == 0:
        print('Par')
    else:
        print('Impar')

par(num_par)
````

### **EJEMPLO 3:**

Función para calcular el factorial de un número:

````python
# difinir función

num_fact = int(input('Ingresa un número: '))

def factorial(numero):
    if numero == 0:
        return 1
    else:
        return numero * factorial(numero - 1)

factorial(num_fact)
````

**3. Define una función que imprima todos los contactos de la agenda:**

   * Crea una función que realice un ciclo for que recorra y muestre cada contacto de la agenda.
   * Dentro del ciclo, muestra el número de contacto por cada iteración.

````python
# Crear función mostrar agenda
def get_max_size(_list, _key):
  newList = [len(item[_key]) for item in _list] if type(_list) == list else [0]
  return max(newList)

def see_all_records(_planner = planner):
  if (len(_planner) == 0):
    print('****Agenda vacía****')
    return

  headers = _planner[0].keys()
  maxSizes = {}
  headersStr = ''
  rows = ''

  for column in headers:
    maxSizes[column] = get_max_size(_planner, column)
    headersStr += f'{column.capitalize()} {(maxSizes[column] - len(column)) * ' '}'

  for contact in _planner:
    for column in headers:
      rows += f'{contact[column]} {(maxSizes[column] - len(contact[column])) * ' '}'
    rows += '\n'

  print(headersStr)
  print(rows)

see_all_records()
````

**5. Define una función para consultar un contacto de la agenda:**

   * Crea una función que pida al usuario que ingrese el nombre del contacto a consultar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces imprime el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Crear función consultar contacto

def find_contact(fullname):
  contactFound = None
  index = 0
  for contact in planner:
    if contact['fullname'] == fullname:
      contactFound = contact
      break
    index += 1
  return {
      "contact": contactFound,
      "index": index
  }

def show_contact():
  fullname = input("Ingresa el nombre: ")
  contactFound = find_contact(fullname)['contact']

  if contactFound != None:
    see_all_records([contactFound])
  else:
    print('Contacto no encontrado')

show_contact()
````

**4. Define una función para crear un contacto de la agenda:**

   * Crea una función que pida al usuario que ingrese el nombre, teléfono y correo para un contacto.
   * Guarda estos datos en un diccionario llamado contacto.
   * Imprime el mensaje "Contacto creado con éxito".

````python
# Crear función crear contacto
def add_contact():
  fullname = input('Ingresa el nombre: ')
  email = input('Ingresa el email: ')
  cellphone = input('Ingresa el celular: ')
  contactFound = find_contact(fullname)

  if is_empty(fullname) or is_empty(email) or is_empty(cellphone):
    print('No se aceptan campos vacios')
    return

  if contactFound['contact'] == None:
    contact = {
        'fullname': fullname,
        "email": email,
        "cellphone": cellphone
    }
    planner.append(contact)
    print('Contacto creado con éxito')
  else:
    print(f'{fullname} ya existe dentro de la agenda, intenta con uno nuevo')
````

   * Crea una función para cada opción a escojer, modificando ya sea el nombre, el teléfono o el correo del contacto encontrado por nuevos registros.
   * Por cada registro nuevo ingresado, imprime el contacto para visualizar los cambios.

````python
# Crear función actualizar nombre para actualizar contacto
def update_contact_fullname(fullname, newFullname):
  if (is_empty(newFullname)):
    print('El nombre no puede ser vacio')
    return

  contactFound = find_contact(fullname)
  if contactFound['contact'] != None:
    print(f'({fullname}) actualizado por ({newFullname})')
    planner[contactFound['index']]['fullname'] = newFullname
  else:
    print("Nombre no encontrado")
````

````python
# Crear función actualizar telefono para actualizar contacto
def update_contact_cellphone(fullname, newCellphone):
  if (is_empty(newCellphone)):
    print('El celular no puede ser vacio')
    return

  contactFound = find_contact(fullname)
  if contactFound['contact'] != None:
    cellphone = contactFound['contact']['cellphone']
    print(f'({cellphone}) actualizado por ({newCellphone})')
    planner[contactFound['index']]['cellphone'] = newCellphone
  else:
    print("Celular no encontrado")
````

````python
# Crear función actualizar correo para actualizar contacto
def update_contact_email(fullname, newEmail):
  if (is_empty(newEmail)):
    print('El correo no puede ser vacio')
    return

  contactFound = find_contact(fullname)
  if contactFound['contact'] != None:
    email = contactFound['contact']['email']
    print(f'({email}) actualizado por ({newEmail})')
    planner[contactFound['index']]['email'] = newEmail
  else:
    print("Correo no encontrado")
````

   * Crea una función que integre las funciones creadas anteriormente a través de un ciclo while para un menú de actualizar un contacto.
   * Pide al usuario que ingrese el nombre del contacto a actualizar.
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Opción no encontrada".
   * Dada la opción escojida, ingresa a una función definida con anterioridad que corresponda a su operación a realizar.
   * Define en la última opción un `break` para salir del menú.

````python
# Crear función actualizar contacto
def show_update_contact():
  fullname = input("Ingresa el nombre del contacto: ")
  contactFound = find_contact(fullname)

  if contactFound['contact'] is not None:
    show_update_menu()
    optionSelected = int(input('Ingresa una opción valida: '))

    if (optionSelected > 0 and optionSelected < 5):
      while (optionSelected != 4):
        if (optionSelected == 1):
          newFullname = input('Ingresa el nuevo nombre: ')
          update_contact_fullname(fullname, newFullname)
        elif (optionSelected == 2):
          newCellphone = input('Ingresa el nuevo celular: ')
          update_contact_cellphone(fullname, newCellphone)
        elif (optionSelected == 3):
          newEmail = input('Ingresa el nuevo correo: ')
          update_contact_email(fullname, newEmail)
        show_update_menu()
        optionSelected = int(input('Ingresa una opción valida: '))
      print('Saliendo del menu actualizar...')
    else:
      print('Opción no encontrada, vuleve a intentar.')
  else:
    print("Contacto no encontrado")

show_update_contact()
````

**7. Define una función para eliminar un contacto de la agenda:**

   * Crea una función que pida al usuario que ingrese el nombre del contacto a eliminar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces elimina el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Crear función eliminar contacto
def delete_contact():
  fullname = input('Ingresa el nombre del contacto a eliminar: ')
  contactFound = find_contact(fullname)
  if contactFound['contact'] != None:
    planner.pop(contactFound['index'])
    print(f'Contacto "{fullname}" eliminado con exito')
  else:
    print(f'Contacto "{fullname}" no encontrado')
````

**8. Junta todas las funciones del menú CRUD dentro de una función:**

   * Crea una función crear un menú a través de un ciclo while.
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Opción no encontrada".
   * Dada la opción escojida, ingresa a una función definida con anterioridad que corresponda a su operación a realizar.
   * Define en la última opción un `break` para salir del menú.

````python
# Crear función menú agenda
def executePlannerCRUD():
  show_planner_menu()
  optionSelected = int(input("Ingresa una opción del menu: "))

  if (optionSelected > 0 and optionSelected < 7):
    while (optionSelected != 6):
      if (optionSelected == 1):
        see_all_records()
      elif (optionSelected == 2):
        add_contact()
      elif (optionSelected == 3):
        show_contact()
      elif (optionSelected == 4):
        show_update_contact()
      elif (optionSelected == 5):
        delete_contact()
      show_planner_menu()
      optionSelected = int(input('Ingresa una opción valida: '))
    print('Saliendo del menu actualizar...')
  else:
    print('Opción no encontrada, vuleve a intentar.')


executePlannerCRUD()
````
